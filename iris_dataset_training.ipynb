```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_excel("iristrain.xlsx")
data
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1]
y
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
prediction=model.predict([[5.1	,3.8	,1.5	,0.3]])
prediction
```
